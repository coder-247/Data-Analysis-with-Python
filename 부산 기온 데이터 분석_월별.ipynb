```python
import csv
import matplotlib.pyplot as plt
f = open('Busan.csv','r', encoding='cp949')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)     
plt.show()


```
